### Regresion Lineal en python con sckit-learn
El modelo selecciona las variables predictorias. Tu solo observa

```python
from sklearn.feature_selection import RFE #Resourced Fisher Elimination
from sklearn.svm import SVR #super vector machine
```

```python
data_full.head()
```

```python
feature_cols = ["No of Vehicle", "Vacations/Year","Age",]#columnas que en principio seran predictoras
feature_cols
```

```python
X = data_full[feature_cols] #predictoras
Y = data_full["Salary"]
```

```python
estimator = SVR(kernel="linear")
selector = RFE(estimator,2, step=1) #le indico que quiero dos varaibles
selector = selector.fit(X,Y)
```

```python
selector.support_
```

```python
selector.ranking_
```
