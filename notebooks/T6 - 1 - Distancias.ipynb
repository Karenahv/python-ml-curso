# Distancias

```python
from scipy.spatial import distance_matrix
import pandas as pd
```

```python
data = pd.read_csv("../datasets/movies/movies.csv", sep=";")
data
```

```python
movies = data.columns.values.tolist()[1:]
movies #quite el user ID
```

```python
dd1 = distance_matrix(data[movies], data[movies], p=1)
dd2 = distance_matrix(data[movies], data[movies], p=2)
dd10 = distance_matrix(data[movies], data[movies], p=10)#entre mas grande el valor, mas pequeña la distancia
#mas se juntan los valores
```

```python
#para pasar los datos a dataframe
def dm_to_df(dd,col_name):
    return pd.DataFrame(dd, index=col_name, columns=col_name)
```

```python
dm_to_df(dd1,data["user_id"])
```

Los valores mas pequeños, **alrededor del 0 por ejemplo significa que son usuarios muy parecidos**
**Entre mas alto el valor menos se parecen**

```python
dm_to_df(dd10,data["user_id"])
```

## Clustering Jerarquico en Python
Vamos a realizar ese Clustering y esos graficos con Python :D

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
movies #recordemos los datos que teniamos
```

```python
data[movies] #devuelve valoracion de peliculas
```

### linkage()
Calcula el enlace del dataet con el metodo que digamos.

**Ward** -> mar rararo, minimiza los cuadrados de las varianzas.
* Podemos usar cualquier otro

```python
Z = linkage(data[movies], "ward")
Z
```

Vamos a crear el Dendograma de forma muy genial

```python
plt.figure(figsize=(25,10))
plt.title("Dendograma jerárquico para el Clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z) #quite la varaible leaf_rotation=90. , por que me mostraba los datos al reves
#quite la varaible leaf_font_size=10.0 por que en mi caso se ve mejor como esta
plt.show()
```

Podemos pintarlo Horizontal

```python
plt.figure(figsize=(25,10))
plt.title("Dendograma jerárquico para el Clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z, orientation="right") #quite la varaible leaf_rotation=90. , por que me mostraba los datos al reves
#quite la varaible leaf_font_size=10.0 por que en mi caso se ve mejor como esta
plt.show()
```

Usando el enlace promedio (otro metodo de medicion)

```python
Z = linkage(data[movies], "average")
Z
plt.figure(figsize=(25,10))
plt.title("Dendrograma jerárquico para el Clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z)
plt.show()
```

Son algo diferentes en el "ward" unimos 3 con 8 y en este ultimo 8 se une a 0 y 9

```python
data[movies]
```

## Muy importante que tenga sentido desde el negocio
Como: **¿Por que el #2 ha sido tirado fuera de los cluster?** es atipico, no es normal esas calificaciones que dio
* Los datos se unen entre usuarios que hacen califiaciones parecidas
* Por ejemplo el #3 con el #8 se unieron en "ward" pero hay una distancia considerable entre ellos

**Probemos el enlace completo**
Los reusltados son similares a los de "ward" (por no decir iguales)

```python
Z = linkage(data[movies], "complete")
Z
plt.figure(figsize=(25,10))
plt.title("Dendrograma jerárquico para el Clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z)
plt.show()
```

```python
Z = linkage(data[movies], method="single", metric="cosine")
Z
plt.figure(figsize=(25,10))
plt.title("Dendrograma jerárquico para el Clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z)
plt.show()
```

Podemos aportar ademas el parametro de **metrica** al definir linkage()
* Por defecto se usa la **Euclidea**
* Las siguientes pueden ser las metricas:

The distance function can be ‘braycurtis’, ‘canberra’, ‘chebyshev’, ‘cityblock’, ‘correlation’, ‘cosine’, ‘dice’, ‘euclidean’, ‘hamming’, ‘jaccard’, ‘kulsinski’, ‘mahalanobis’, ‘matching’, ‘minkowski’, ‘rogerstanimoto’, ‘russellrao’, ‘seuclidean’, ‘sokalmichener’, ‘sokalsneath’, ‘sqeuclidean’, ‘yule’.
