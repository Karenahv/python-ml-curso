Vamos a usarla en un problema que ira creciendo en dificultad poco a poco

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

import seaborn as sns; sns.set() #otra forma de llevar a cabo las representaciones grafica
```

```python
from sklearn.datasets.samples_generator import make_blobs #(traduce= hacer burbujas
```

```python
X, Y = make_blobs(n_samples=50, centers=2, random_state=0, cluster_std=0.6) 
#centers= centros de separacion de los datos
#random state= para reproducir resultado
#cluster_std = desviacion estandar
```

```python
plt.scatter(X[:,0], X[:,1], c = Y, s= 50, cmap="autumn")
```

```python
xx = np.linspace(-1, 3.5)
plt.scatter(X[:,0], X[:,1], c = Y, s = 50, cmap="autumn")
plt.plot([0.5], [2.1], 'x', color="blue", markeredgewidth=2, markersize=10)

#vamos a crear un conjunto de reglas, las cuales podrian ser todas muy validas
for a, b in [(1,0.65), (0.5,1.6), (-0.2,2.9)]:#para estas tres rectas
    yy = a * xx + b
    plt.plot(xx,yy,"-k")
plt.xlim(-1,3.5)
```

Dependiendo de la linea (clasificador) el punto podria ser de uno u otro **¡Este es el problema del separador! Cual elegir?**

## Maximización del margen

```python
xx = np.linspace(-1, 3.5)
plt.scatter(X[:,0], X[:,1], c = Y, s = 50, cmap="autumn")
plt.plot([0.5], [2.1], 'x', color="blue", markeredgewidth=2, markersize=10)

#vamos a crear un conjunto de reglas, las cuales podrian ser todas muy validas
for a, b, d in [(1,0.65,0.33), (0.5,1.6,0.55), (-0.2,2.9,0.2)]:#para estas tres rectas, estso valores son dados por el profe
    yy = a * xx + b
    plt.plot(xx,yy,"-k")
    plt.fill_between(xx, yy-d, yy+d, edgecolor="none", color="#BBBBBB", alpha=0.4)
plt.xlim(-1,3.5)
```

Una SVM lo que hace es elegir la que tenga el margen maximizado (en esta caso queremos decir que el pasillo mas ancho es el mejor)
